```python
import pandas as pd

def create_location_table(area_locations):
    """Creates a location table from the area_locations dictionary and saves it to a CSV file."""

    locations = []
    location_id = 1  # Start location ID from 1

    for area_code, area_locations_list in area_locations.items():
        for location_name, latitude, longitude in area_locations_list:
            locations.append([location_id, area_code, location_name, latitude, longitude])
            location_id += 1  # Increment location ID

    location_df = pd.DataFrame(locations, columns=['LocationID', 'AreaCode', 'LocationName', 'Latitude', 'Longitude'])

    # Save the DataFrame to a CSV file
    location_df.to_csv('Location.csv', index=False)  # index=False prevents writing the DataFrame index to the CSV

    return location_df


location_df = create_location_table(area_locations)
location_dict = location_df.set_index('LocationID').to_dict('index')
```
